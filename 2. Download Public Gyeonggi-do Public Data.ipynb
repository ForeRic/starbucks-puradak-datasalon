```python
# select only for 2021 data (Jan - Mar)
find_row = sigun_pop_df_remove_do.loc[(sigun_pop_df['Year']) == '2021']
find_row
```

```python
# 두개를 밑에 병합시켜서 가공해야함
only_march_sigun = only_march.loc[(only_march['Sigun']) == '시군']
#only_march_gu = only_march.loc[(only_march['Sigun'] == '구')]
sigun = only_march_sigun.reset_index(drop=True)
sigun
#gu = only_march_gu.reset_index(drop=True)
#add_sigun_gu = pd.concat([sigun,gu])
#sgg=add_sigun_gu.reset_index(drop=True)
#sgg.head()
```
