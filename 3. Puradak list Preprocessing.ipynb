## 3. 푸라닭 리스트 전처리

```python
# 1번 크롤링해서 얻은 데이터 저장된 엑셀파일 불러오기
import pandas as pd
puradak_gyeunggi = pd.read_excel('./files/puradak_list.xlsx', header=0, engine='openpyxl')
puradak_gyeunggi
```

```python
# 새로운 컬럼 추가하기 
sigun_names = []
for address in puradak_gyeunggi['주소']:
    sigun = address.split()[1] #  주소를 분리해서 인덱스처럼 나누고 거기서 선택     
    sigun_names.append(sigun)
    
#sigun_names
puradak_gyeunggi['시군명'] = sigun_names #  새칼럼형성 
puradak_gyeunggi.head()
```

```python
# 원래 있던 엑셀파일에 새로운 컬럼 추가해서 업데이트
puradak_gyeunggi.to_excel('./files/puradak_list.xlsx', index =False, engine='openpyxl')
```
