## 5. 푸라닭 매장 분포 시각화 (folium)

```python
import pandas as pd
import folium
import json
```

```python
puradak_list = pd.read_excel('./files/puradak_list.xlsx', header=0, engine='openpyxl')
puradak_list
```

```python
# 6.3-3 folium을 이용한 지도 생성 - 경기도가 지도에서 어디쯤에 있는지 표시
puradak_map = folium.Map(
    location=[37.587218, 127.232031],
    tiles='Stamen Terrain',
    zoom_start=8
)

# Add code : :현재:현재위치표시  
folium.Marker(
  location=[37.587218, 127.232031],
  popup='Here',
  icon=folium.Icon(color='red',icon='star')
).add_to(puradak_map)

puradak_map
```

```python
# 경기도에 위치한 푸라닭 매장 위치 표시
puradak_map2 = folium.Map(
    location=[37.587218, 127.232031],
    tiles='CartoDB positron',
    zoom_start=9
)

for idx in puradak_list.index:
    lat = puradak_list.loc[idx, '위도']
    lng = puradak_list.loc[idx, '경도']

    folium.CircleMarker(
        location=[lat, lng],
        fill = True, 
        fill_color='red', 
        fill_opacity=1, # 0.5 : faint
        color='black', 
        weight=0.5,
        radius=3
    ).add_to(puradak_map2)

puradak_map2
```
